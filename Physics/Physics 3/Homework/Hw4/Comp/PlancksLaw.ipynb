```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from __future__ import division
```

```python
l= np.linspace(0.1, 5, 1000)
t = [1000*i for i in range(1, 5)]
hc = 1.23984
kb = 8.617333 * 10**-5
u = []
fig = plt.figure(0)
ax = fig.add_axes([0,0,1,1], title='Planck\'s Law Dependence on Temperature')
for i in range(4):
    u.append(8*np.pi*hc/(l**5 * (np.exp(hc/(kb*l*t[i])) - 1)))
    ax.plot(l, u[i], label='T={}'.format(t[i]))
    lmax = int(np.argmax(u[i]))
    ax.annotate("u = {:.4f}".format(u[i][lmax]), (l[lmax]+0.1,u[i][lmax]))
    ax.legend()
plt.savefig("fig1")
```

```python
l= np.linspace(0.1, 50, 1000)
t = [100*i for i in range(1, 5)]
fig = plt.figure(1)
ax = fig.add_axes([0, 0, 1, 1], title='Planck\'s Law Dependence on Temperature')
u = []
for i in range(4):
    u.append(8 * np.pi * hc / (l ** 5 * (np.exp(hc / (kb * l * t[i])) - 1)))
    ax.plot(l, u[i], label='T={}'.format(t[i]))
    lmax = int(np.argmax(u[i]))
    ax.annotate("u = {:.3e}".format(u[i][lmax]), (l[lmax] + 0.1, u[i][lmax]))
    ax.legend()
plt.savefig("fig")
```
