```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import sympy as sp
from __future__ import division
```

```python
l= sp.symbols('l')
t = [1000*i for i in range(1, 5)]

hc = 1.23984
kb = 8.617333 * 10**-5
u = []
plots = []
for i in range(4):
    u.append(8*np.pi*hc/(l**5 * (sp.exp(hc/(kb*l*t[i])) - 1)))
    plots.append(sp.plot(u[i], (l, 0.001, 5), nb_of_points=1000, ylabel="$u$".format(t[0]), show=False, xlabel="$\lambda$"))
    if i != 0:
        plots[0].extend(plots[i])
plots[0].show()
t = [100*i for i in range(1, 5)]
u = []
plots = []
for i in range(4):
    u.append(8*np.pi*hc/(l**5 * (sp.exp(hc/(kb*l*t[i])) - 1)))
    plots.append(sp.plot(u[i], (l, 0.1, 50), nb_of_points=1000, ylabel="$u$".format(t[0]), show=False, xlabel="$\lambda$"))
    if i != 0:
        plots[0].extend(plots[i])
plots[0].show()
```
