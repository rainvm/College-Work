# 3
The equation which describes the wavelength of an emitted photon is $\frac{1}{\lambda} = \frac{ke^2}{2a_0hc}\left(\frac{1}{n_f^2}-\frac{1}{n_i^2}\right)$. This equation does not depend on the mass of the nucleus, and therefore the emission spectrum should also not depend on the mass of the nucleus.
