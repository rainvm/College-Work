```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
h1 = pd.read_csv("hydrogen_1.csv")
h2 = pd.read_csv("hydrogen_2.csv")
d1 = pd.read_csv("deuterium_1.csv")
d2 = pd.read_csv("deuterium_2.csv")
neon = pd.read_csv("neon_1.csv")
```

# 2.

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.85,0.8])
ax.plot(d1.Wavelength, d1.Intensity, label="Unsaturated Deuterium")
ax.plot(h1.Wavelength, h1.Intensity, label="Unsaturated Hydrogen")
ax.legend()
```

# 6.

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.85,0.8])
ax.plot(neon.Wavelength, neon.Intensity, label="Neon")
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.85,0.8])
ax.plot(d2.Wavelength, d2.Intensity, label="Saturated Deuterium")
ax.plot(h2.Wavelength, h2.Intensity, label="Saturated Hydrogen")
ax.legend()
```
