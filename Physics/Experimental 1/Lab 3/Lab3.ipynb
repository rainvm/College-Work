```python
import polars as po
import numpy as np
import matplotlib.pyplot as plt

file_names = ['low-pass.csv', 'high-pass.csv', 'network-analyzer.csv', 'high-pass-net.csv', 'rlc-net.csv']
plot_names = ['Low Pass Filter', 'High Pass Filter']
frames = []
for i in file_names:
    frames.append(po.read_csv(i))

for i in range(2):
    plt.plot(frames[i][['Time (s)']], frames[i][['Channel 1 (V)']], label='Input Voltage')
    plt.plot(frames[i][['Time (s)']], frames[i][['Channel 2 (V)']], label='Output Voltage')
    plt.legend()
    plt.title(plot_names[i])
    plt.xlabel("Time (S)")
    plt.ylabel("Voltage (V)")
    plt.show()
    
```
