```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Filter(f, R1, R2, C1, C2):
    Z1 = 1/(2j*np.pi*f*C1)
    Z2 = 1/(2j*np.pi*f*C2)
    Zeq = 1/(1/Z1 + 1/(Z2+R2))
    v1 = Zeq/(R1 + Zeq)
    out = v1*(R2/(R2 + Z2))
    return out
f = np.linspace(5, 10000, 100000)
g = Filter(f, 1e4, 1e4, 100e-9, 100e-9)
plt.plot(f, g)
plt.xscale("log")
print(f[np.argmax(g)])
```
