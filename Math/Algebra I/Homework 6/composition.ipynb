```python
import numpy as np
autz24 = [i for i in range(24) if np.gcd(i, 24)==1]
for i in autz24:
    line = f"\(f_{{{i}}}\) & "
    for j in autz24:
        line += f"\(f_{{{i*j%24}}}\) & "
    print(line[:-2] + "\\\\")
```

```python
H = [0, 4, 1j, 4 + 1j]
for r in range(7):
    line = "\\{"
    for i in H:
        if np.imag(i) != 0:
            line += "s"
        line += f"r^{{{int((np.real(i) + r)%8)}}}, "
    print(line[:-2] + "\\} \\\\")
```
